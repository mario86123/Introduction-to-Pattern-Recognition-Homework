```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train_df = pd.read_csv("train_data.csv")
x_train, y_train = train_df['x_train'], train_df['y_train']
print(train_df)
type(train_df)
```

```python
plt.plot(x_train, y_train, '.')
```

```python
weight = 0
bias = 0
iter_time = 30
learning_rate = 0.1
batch = 5
loss = [0] * iter_time
axis = [t for t in range(iter_time)]
```

```python
for i in range(iter_time):
    print(i)
    
    loss[i] = (np.sum(np.power(y_train - (bias + weight * x_train), 2))) / 500
    print("loss[i]: ", loss[i])
    
    w_grad = np.sum((-2) * (y_train - (bias + weight * x_train)) * x_train) / 500
    b_grad = np.sum((-2) * (y_train - (bias + weight * x_train))) / 500
    
    weight = weight - learning_rate * w_grad
    bias = bias - learning_rate * b_grad
    print("weight: ", weight)
    print("bias: ", bias)
    
    x = np.linspace(-3, 3, 100)
    y = weight * x + bias
    plt.plot(x, y, 'r')
    
plt.plot(x_train, y_train, '.')
```

```python
plt.plot(axis, loss)
```
