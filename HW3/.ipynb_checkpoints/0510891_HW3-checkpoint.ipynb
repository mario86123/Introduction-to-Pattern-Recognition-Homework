## Question 1
Gini Index or Entropy is often used for measuring the “best” splitting of the data. Please compute the Entropy and Gini Index of provided data. Please use the formula from page 666 on the textbook

```python
def gini(sequence):
    num_of_class_one = 0
    size = sequence.size
    if size == 0:
        return 0
    for i in sequence:
        if i == 1:
            num_of_class_one += 1
    proportion = num_of_class_one / size
    return 1 - proportion ** 2 - (1 - proportion) ** 2

def entropy(sequence):
    num_of_class_one = 0
    size = sequence.size
    if size == 0:
        return 0
    for i in sequence:
        if i == 1:
            num_of_class_one += 1
    proportion = num_of_class_one / size
    return - proportion * math.log(proportion, 2) - proportion * math.log(proportion, 2)
```

```python
# 1 = class 1,
# 2 = class 2
data = np.array([1,2,1,1,1,1,2,2,1,1,2])
```

```python
print("Gini of data is ", gini(data))
```

```python
print("Entropy of data is ", entropy(data))
```

## Question 2
Implement the Decision Tree algorithm (CART, Classification and Regression Trees) and trained the model by the given arguments, and print the accuracy score on the test data. You should implement two arguments for the Decision Tree algorithm
1. **Criterion**: The function to measure the quality of a split. Your model should support “gini” for the Gini impurity and “entropy” for the information gain. 
2. **Max_depth**: The maximum depth of the tree. If Max_depth=None, then nodes are expanded until all leaves are pure. Max_depth=1 equals to split data once


```python
class DecisionTree():
    def __init__(self, criterion='gini', max_depth=None):
        return None
```

```python
def split(index, value, dataset):
    left, right = list(), list()
    for row in dataset:
        if row[index] < value:
            left.append(row)
        else:
            right.append(row)
    return left, right

def total_gini(groups):
    g, y = [], []
    for group in groups:
        for i in range(len(group)):
            y.append(group[i][-1] + 1)
        g.append(gini(np.array(y)))
    p = len(groups[0]) / (len(groups[0]) + len(groups[1]))
    return  p * g[0] + (1 - p) * g[1]
    
def get_split(dataset):
    best_index, best_compare_num, best_gini, best_groups = 0, 0, 999, None
    for index in range(len(dataset[0]) - 1):
        i = 0
        for row in dataset:
            groups = split(index, row[index], dataset)
            gini = total_gini(groups)
            i += 1
            print('%d < %.3f Gini=%.3f %d' % ((index+1), row[index], gini, i))
            if gini < best_gini:
                best_index, best_compare_num, best_gini, best_groups = index, row[index], gini, groups
    return {'index':best_index, 'value':best_gini, 'groups':best_groups}

s = get_split(raw_x)
print('Split: [%d < %.3f]' % ((feature_names['index']), s['value']))
```
